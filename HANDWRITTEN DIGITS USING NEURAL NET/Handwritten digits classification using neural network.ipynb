```python
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
#LOAD HANDWRITTEN DIGITS DATASET
(x_train,y_train),(x_test,y_test)=keras.datasets.mnist.load_data()
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
x_train[0].shape
```

```python
x_train[0]
```

```python
plt.matshow(x_train[0])
```

```python
y_train[0]
```

```python
#SCALING---IT IS THE TECHIQUE WHICH IMPROVES THE ACCURASY OF MECHINE LEARNING MODELS
x_train=x_train/255
x_test=x_test/255
```

```python
x_train[0]
```

```python
x_train_flattened=x_train.reshape(len(x_train),28*28)
x_test_flattened=x_test.reshape(len(x_test),28*28)
```

```python
x_train_flattened.shape
```

```python
x_train_flattened[0]
```

```python
#CREATE SIMPLE NEURAL NETWORK
model=keras.Sequential([
    keras.layers.Dense(10,input_shape=(784,),activation='sigmoid')
])
#OPTIMIZER ALLOWS YOU TO TRIQIN EFFICIENCY.
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
model.fit(x_train_flattened, y_train, epochs=5)
```

```python
#EVALUATING ACCURACY OF A DATASET
model.evaluate(x_test_flattened,y_test)
```

```python
#SAMPLE PREDICTION AND IT WILL GIVE THE PREDICTION OF 10 CLASSES
y_predicted=model.predict(x_test_flattened)
y_predicted[0]
```

```python
plt.matshow(x_test[0])
```

```python
np.argmax(y_predicted[0])
```

```python
y_predicted_labels=[np.argmax(i) for i in y_predicted]
```

```python
#IT SHOWS THE VALUE FROM THE FIRST 5 INDEXES

y_predicted_labels[:5]
```

```python
cm=tf.math.confusion_matrix(labels=y_test,predictions=y_predicted_labels)
cm
```

```python
#VISUALIZING THE CNFUSION MATRIX
import seaborn as sn
plt.figure(figsize=(10,7))
sn.heatmap(cm,annot=True,fmt='d')
plt.xlabel('predicted')
plt.ylabel('Truth')
```

```python
#using hidden layer
model=keras.Sequential([
    
    keras.layers.Dense(100,input_shape=(784,),activation='relu'),
    keras.layers.Dense(10,activation='sigmoid')

])

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
model.fit(x_train_flattened, y_train, epochs=5)
```

```python
model.evaluate(x_test_flattened,y_test)
```

```python
y_predicted=model.predict(x_test_flattened)
y_predicted_labels=[np.argmax(i) for i in y_predicted]
cm=tf.math.confusion_matrix(labels=y_test,predictions=y_predicted_labels)


plt.figure(figsize=(10,7))
sn.heatmap(cm,annot=True,fmt='d')
plt.xlabel('predicted')
plt.ylabel('Truth')
```

```python
model=keras.Sequential([
    keras.layers.Flatten(input_shape=(28,28)),
    keras.layers.Dense(100,activation='relu'),
    keras.layers.Dense(10,activation='sigmoid')

])

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
model.fit(x_train, y_train, epochs=10)
```

```python
model.evaluate(x_test,y_test)
```
